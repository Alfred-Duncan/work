```python
import pandas as pd
import numpy as np
from pyecharts import options as opts
from pyecharts.charts import Map
import random
def create_world_map():
    (   
        Map()
        .add(
            series_name="lab",
            data_pair=c,
            maptype="world",
        )
        .set_global_opts(
            title_opts=opts.TitleOpts(title="lab"),
            visualmap_opts=opts.VisualMapOpts(max_=20000, is_piecewise=True),
        )
        .set_series_opts(
            label_opts=opts.LabelOpts(is_show=False, color="blue")
        )
        .render("lab.html")
    )
create_world_map()
##去往项目同级目录下查看生成的html
```
