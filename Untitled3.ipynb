```python
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt
data_x = np.arange(7)
con1 = [119790.59, 2505227.53, 2090037.5, 2162194.14, 2815830.87, 3045066.26, 2785745.23]
con2 = [70636.04, 3280031.09, 3142878.76, 3189964.83, 3683943.5, 3630726.83, 4280293.27]
con3 = [187932.43, 2222094.1, 2480200.08, 2373739.41, 2845497.49, 3201003.74, 2855998.52]
bar_width = 0.1
tick_label = ["Day1", "Day2", "Day3", "Day4", "Day5", "Day6", "Day7"]
plt.bar(data_x,
        con1,
        bar_width,
        align="center",
        color="y",
        label="ConcourseA",
        alpha=0.3)
plt.bar(data_x + bar_width,
        con2,
        bar_width,
        align="center",
        color="b",
        label="ConcourseB",
        alpha=0.3)
plt.bar(data_x + bar_width + bar_width,
        con3,
        bar_width,
        align="center",
        color="r",
        label="ConcourseC",
        alpha=0.3)
plt.xlabel("Data")
plt.ylabel("Sold")
plt.xticks(data_x + bar_width / 2, tick_label)
plt.legend()
plt.show()
```
