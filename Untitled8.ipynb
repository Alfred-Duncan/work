```python
import numpy as np
import pandas as pd
from pyecharts.globals import ThemeType
import pyecharts.options as opts
from pyecharts.charts import ThemeRiver
df_ThemeRiver = pd.read_csv("df1.csv",encoding='utf-8',engine='python')
#df_ThemeRiver['hour2'] = df_ThemeRiver.trx_datime_c1.str.split(":").str[0]
df_ThemeRiver_country = df_ThemeRiver[ (df_ThemeRiver.concourse_c1 == "Concourse A") | (df_ThemeRiver.concourse_c1 == "Concourse B") | (df_ThemeRiver.concourse_c1 == "Concourse C")]
df_ThemeRiver_country = df_ThemeRiver_country[ ["data","time","concourse_c1"] ]
#df_ThemeRiver_country = df_ThemeRiver_country[ ["hour2","time","concourse_c1"] ]
#list(df_ThemeRiver_country["concourse_c1"].unique())
#df_ThemeRiver_country.values.tolist()
c_ThemeRiver = (
    ThemeRiver(init_opts=opts.InitOpts(width="1000px", height="600px",theme = ThemeType.CHALK))
    .add(
        series_name=list(df_ThemeRiver_country["concourse_c1"].unique()),
        data=df_ThemeRiver_country.values.tolist(),
        singleaxis_opts=opts.SingleAxisOpts(
            pos_top="50", pos_bottom="50", type_="time"
        ),
    )
    .set_global_opts(
        tooltip_opts=opts.TooltipOpts(trigger="axis", axis_pointer_type="line"),
        title_opts=opts.TitleOpts(pos_bottom = "85%", pos_right = "20%")
    )
    .set_series_opts(label_opts=opts.LabelOpts(is_show = 0))
    #.render("theme_river.html")
)
c_ThemeRiver.render_notebook()
```
